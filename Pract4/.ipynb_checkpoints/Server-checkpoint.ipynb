### MULTITHREADING-1

```python
import socket
import sys
import threading
import time

def worker_routine(thread_name, connection):
       
    print("Thread "+ thread_name + " started")
    msg_rx = ""
    while msg_rx != "exit":
        try:
            msg_rx = connection.recv(1024).decode()
            if msg_rx != "":
                print("RX: " + msg_rx)
            msg_tx = msg_rx
            connection.sendall(msg_tx.encode()) # Send received message back to client
        except:
            break
    
    print("Thread "+ thread_name + " ended")
    connection.close() # Close connection

def start_thread(thread_name, connection):
    reader = threading.Thread(target=worker_routine, args=(thread_name, connection))

    reader.start()
    
    
    
port = input("Input server port: ")

# Check if port is a number
while port.isnumeric() == False:
    port = input("Port must be numeric: ")

port = int(port)
    
# Check if port is valid and create socket
try:
        server_address = ("localhost", port)
        sock = socket.socket(family=socket.AF_INET, type=socket.SOCK_STREAM)
        sock.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
        sock.bind(server_address)
        sock.listen(1)
        
except Exception as e:
        print("Invalid port")
        print(e)
    

port = int(port)

iteration = 0 # To identify threads

try:
    while True:
        connection, client_address = sock.accept() # Receive new connection and give it to a new thread
        start_thread(str(iteration), connection)

        iteration = iteration + 1
            
except Exception as e:
    print(e)
    
```

```python
import socket
import sys
import threading
import time

def worker_routine(thread_name, connection):
       
    msg_rx = ""
    while msg_rx != "exit":
        try:
            msg_rx = connection.recv(1024).decode()
            if msg_rx != "":
                print_lock.acquire() ## Lock prints
                print("Received: -->  " + msg_rx)
                print("Sending data back to the client")
                print_lock.release() ## Unlock prints
            msg_tx = msg_rx
            connection.sendall(msg_tx.encode())
        except:
            break
    
    connection.close()

def start_thread(thread_name, connection):
    reader = threading.Thread(target=worker_routine, args=(thread_name, connection))

    reader.start()
    
    
    
print_lock = threading.Lock()
port = input("Input server port: ")
# Check if port is a number
while port.isnumeric() == False:
    port = input("Port must be numeric: ")

port = int(port)
    
try:
        server_address = ("localhost", port)
        sock = socket.socket(family=socket.AF_INET, type=socket.SOCK_STREAM)
        sock.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
        sock.bind(server_address)
        sock.listen(1)
        print("Starting server on port (localhost, "+str(port)+") ")
        
except Exception as e:
        print("Invalid port")
        print(e)
    

port = int(port)

iteration = 0

try:
    while True:
        print_lock.acquire()
        print("Waiting for a connection")
        connection, client_address = sock.accept()
        print("Connection received from " + str(client_address) )
        print_lock.release()
        start_thread(str(iteration), connection)

        iteration = iteration + 1
            
except Exception as e:
    print(e)
    
```
